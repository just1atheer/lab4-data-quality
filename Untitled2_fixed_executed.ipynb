# Heart/Insurance Data Quality and Preprocessing

This notebook performs basic data quality checks, preprocessing, normalization, standardization, correlation analysis, and PCA on the provided dataset.

## 1) Import libraries and load the dataset

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('insurance.csv')
df.head()
```

## 2) Explore the dataset
We check the structure of the dataset, data types, missing values, and duplicated rows.

```python

print("Shape:", df.shape)
print("\nColumns:", list(df.columns))
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
print("\nDuplicated rows:", df.duplicated().sum())
```

## 3) Select numerical columns
These columns will be used for preprocessing and analysis.

```python

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numerical columns:", list(numeric_cols))
```

## 4) Handle missing values
In this dataset, there are no missing values. However, the following step shows how missing numerical values could be filled using the column mean if needed.

```python

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

print(df.isnull().sum())
```

## 5) Remove duplicated rows
Duplicated rows may affect the quality of analysis, so we remove them.

```python

before_rows = df.shape[0]
df.drop_duplicates(inplace=True)
after_rows = df.shape[0]

print("Rows before removing duplicates:", before_rows)
print("Rows after removing duplicates:", after_rows)
print("Number of removed duplicate rows:", before_rows - after_rows)
```

## 6) Detect and remove outliers using IQR
For each numerical column, we calculate the interquartile range (IQR) and remove values outside the accepted range.

```python

df_no_outliers = df.copy()

for col in numeric_cols:
    Q1 = df_no_outliers[col].quantile(0.25)
    Q3 = df_no_outliers[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_no_outliers = df_no_outliers[(df_no_outliers[col] >= lower) & (df_no_outliers[col] <= upper)]

print("Original shape:", df.shape)
print("Shape after removing outliers:", df_no_outliers.shape)
```

## 7) Visualize outliers with boxplots
Boxplots help us see whether extreme values are present in the numerical features.

```python

for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    plt.boxplot(df[col].dropna(), vert=False)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

## 8) Apply Min-Max normalization
This scales numerical values to a range between 0 and 1.

```python

df_minmax = df_no_outliers.copy()

for col in numeric_cols:
    min_val = df_minmax[col].min()
    max_val = df_minmax[col].max()
    df_minmax[col] = (df_minmax[col] - min_val) / (max_val - min_val)

df_minmax.head()
```

## 9) Apply Z-score standardization
This transforms the data so that each column has mean 0 and standard deviation 1.

```python

df_zscore = df_no_outliers.copy()

for col in numeric_cols:
    df_zscore[col] = (df_zscore[col] - df_zscore[col].mean()) / df_zscore[col].std()

df_zscore[numeric_cols].agg(['mean', 'std']).round(2)
```

## 10) Correlation matrix
This shows the strength of the relationship between the numerical features.

```python

correlation_matrix = df_no_outliers[numeric_cols].corr()
correlation_matrix
```

## 11) Histograms for numerical columns
Histograms help us understand the distribution of each numerical feature.

```python

df_no_outliers[numeric_cols].hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

## 12) PCA (Principal Component Analysis)
PCA reduces the dimensionality of the numerical data while preserving as much variation as possible.

```python

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_no_outliers[numeric_cols])

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)

pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
pca_df.head()
```

## 13) PCA scatter plot
This plot shows the dataset after reducing it to two principal components.

```python

plt.figure(figsize=(7, 5))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection')
plt.show()
```

## 14) Conclusion
- The dataset was loaded and checked successfully.
- There were no missing values, and only duplicate rows needed removal.
- Outliers were handled using the IQR method.
- Numerical features were normalized and standardized.
- Correlation analysis and PCA were applied to better understand the data.
